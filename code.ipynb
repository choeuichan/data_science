# 3600초 (1시간) 이내의 버스 탑승시간이 출근이라 생각

```python
def calculateTime(onTime, offTime, index):
    runClock = []
    for i in index:  
        on = onTime[i].split(':')
        off = offTime[i].split(':')
        onClock = int(on[0]) * 3600 + int(on[1]) * 60 + int(on[2])
        offClock = int(off[0]) * 3600 + int(off[1]) * 60 + int(off[2])
        runClock.append(offClock - onClock)
        
    add_srs = Series(runClock, index)
    return add_srs
```

```python
busNooff = bus.dropna()
busNooff['runClock'] = calculateTime(busNooff['geton_time'], 
                                     busNooff['getoff_time'],
                                    busNooff['geton_time'].index)
```

```python
busNooff = busNooff[ busNooff['runClock'] <=  3600]
busNooff.head()
```
